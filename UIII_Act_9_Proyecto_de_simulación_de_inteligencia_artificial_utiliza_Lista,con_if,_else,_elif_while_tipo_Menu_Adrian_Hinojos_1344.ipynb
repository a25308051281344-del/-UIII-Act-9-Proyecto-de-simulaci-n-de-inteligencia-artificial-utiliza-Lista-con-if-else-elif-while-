```python
# ==========================================
# IA DISEÑADOR VIRTUAL
# Nivel: Principiante
# Usa Lista, if, if else, elif y while
# No usa clases, funciones, for ni try
# ==========================================

# Lista vacía
proyectos = []

print("===================================")
print("     IA DISEÑADOR VIRTUAL")
print("===================================")

while True:

    print("\n--------- MENÚ ---------")
    print("1. Agregar proyecto")
    print("2. Ver proyectos")
    print("3. Analizar proyecto")
    print("4. Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":

        nombre = input("Nombre del proyecto: ")
        tipo = input("Tipo de diseño (Logo, Web, Infografía): ")
        presupuesto = int(input("Presupuesto: "))

        proyectos.append(nombre)

        print("\nProyecto agregado correctamente.")
        print("Total de proyectos:", len(proyectos))

    elif opcion == "2":

        print("\nLISTA DE PROYECTOS")

        if len(proyectos) == 0:
            print("No hay proyectos registrados.")
        else:
            print(proyectos)

    elif opcion == "3":

        print("\nANÁLISIS INTELIGENTE")

        if len(proyectos) == 0:

            print("Primero agrega un proyecto.")

        else:

            tipo = input("Tipo de diseño (Logo, Web, Infografía): ")
            presupuesto = int(input("Presupuesto: "))
            urgente = input("¿Es urgente? (si/no): ")

            if tipo.lower() == "logo" and presupuesto >= 5000:

                print("IA: Recomienda Diseño Premium de Logo.")

            elif tipo.lower() == "web" and presupuesto >= 8000:

                print("IA: Recomienda Página Web Profesional.")

            elif tipo.lower() == "infografía":

                print("IA: Recomienda Diseño Creativo e Informativo.")

            else:

                print("IA: Recomienda Diseño Básico.")

            if urgente == "si":

                print("IA: Proyecto con prioridad alta.")

            else:

                print("IA: Proyecto con prioridad normal.")

    elif opcion == "4":

        print("\nGracias por utilizar IA Diseñador Virtual.")
        break

    else:

        print("\nOpción incorrecta.")
```
